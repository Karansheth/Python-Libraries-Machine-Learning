```python
blood_suger = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_suger, bins=[80, 100, 125, 150], rwidth=0.95, )
```

```python
blood_suger = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_suger, bins=3, rwidth=0.95, histtype='step')
```

```python
blood_suger_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_suger_women = [67, 98, 90, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Suger Range");
plt.ylabel("Total No. of Patients");
plt.title("Blood Suger Analysis");

plt.hist([blood_suger_men, blood_suger_women], bins=[80,100,125,150], rwidth=0.95, color=['green', 'red'],
         label=["Men","Women"], orientation="horizontal")
plt.legend()
```
